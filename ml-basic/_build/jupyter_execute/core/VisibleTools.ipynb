# Matplotlib

准备工作：
- 导入所需要的包
- 数据初始化
- 绘制图像

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
x1 = np.linspace(-5, 5, 100)
y1 = np.sin(x1)
```

调用matplotlib库中的plot函数进行描点绘制图像

```python
plt.plot(x1, y1)
```

此外，可通过调用不同的函数绘制不同的图像

1. 散点图

```python
x2 = np.arange(100)
y2 = x2 * np.random.rand(100)
plt.scatter(x2, y2)
```

2. 直方图

```python
plt.hist(y2, bins=5)
#参数bins为直方图的箱子数量
```

```python
plt.hist(y2, bins=8)
```

3. 柱状图

```python
plt.bar(x2, y2)
```

4. 折线图

```python
plt.plot(x2, y2)
```

### Pandas

准备工作
- 导入数据
- 对DataFrame对象进行操作

```python
import pandas as pd
from sklearn.datasets import load_wine
data = load_wine()
df_X = pd.DataFrame(data.data, columns=data.feature_names)
```

```python
df_X.head()
#默认调用前5行数据
```

```python
df_X.head(8)
```

```python
#将数据的目标变量转换为 pandas 的 DataFrame
df_y = pd.DataFrame(data.target, columns=["kind(target)"])
df_y.head()
```

```python
#将特征值和目标变量的DataFrame合并
df = pd.concat([df_X, df_y], axis=1)
df.head()
```
